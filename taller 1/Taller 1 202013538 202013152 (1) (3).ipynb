# Punto 1. Calentamiento

Construya una función que represente el lanzamiento de un dado de 6 caras. Lance el dado 1000 veces y guarde sus resultado en un vector. Calcule la media empírica y el valor esperado (teórico) de lanzar el mismo dado y compare los resultados. 

````python
!pip install numpy
````

````python

# Importamos libreria Nunpy
import numpy as np

# Fijamos una semilla para replicar resultados
np.random.seed(756432)
````

````python

# Se creo la función "dado" recibe como argumento el numero de intentos y se especifica que el debe retornar un objeto  
def dado(intentos:int) -> list:

    # Lista vacia para almacenar resultados
    resultados = []

    # Un Loop que se repite el numero de intentos
    for i in range(intentos):

        # Se genera un numero aletorio entre el 1 y 6
        intento_dado=np.random.randint(1,7)

        # Se almacena el resultado en la lista
        resultados.append(intento_dado)

    # Se retorna la lista de resultados
    return(resultados)
````

````python

# Se utiliza la función con intentos=1000
mil_intentos=dado(1000)

#Calcular valor esperado e impirmir resultado
valor_esperado= np.sum(range(1,7))/6
print(f"El valor esperado es: {valor_esperado}")

#Calcular media muestral e imprimr resultado
media_muestral=np.mean(mil_intentos)
print(f"La media empírica es: {media_muestral}")

#Comparación de resultados con condicionales.
if media_muestral>valor_esperado:
    print("La media empírica es mayor que el valor esperado teórico")
else:
    print("La media empírica es menor que el valor esperado teórico")
````

### Punto 2.3.

Ahora definamos una función que sea pegar las monas/laminitas de nuestro paquete en nuestro album. 

Para hacer esto suponemos que nuestro `album_i` creado en el punto 2.1. es una lista con todas las monas/laminitas que hacen falta para llenar el album. En ese sentido, "pegar" las monas/laminitas en este consiste en eliminar los números que se tienen en el `sobre_j` del `album_i`.  

````python
# Se define la función "pegar monas" los argumentos son una lista llamada "album" y otra lista llamada "monas".
def pegar_monas(album : list, monas: list) -> list:

    # Asiganamos la una nueva variable al parametro album
    album_actualizado = album

    # Recorremos la lista del que contiene el sobre
    for i in monas:

        # Realizamos un codicional que se activa cuando la mona del sobre no se ha llenado en el album
        if i in album_actualizado:

            # Eliminamos la mona del album
            album_actualizado.remove(i)

    # Retornamos el album actualizado        
    return album_actualizado
````

````python
## Creamos objesto array con la frecuencia, frecuencia relataiva y valores ordenados (menor a mayor) de los resultados de la simulación
valores, frecuencia = np.unique(simulaciones, return_counts=True)
frecuencia_rel= frecuencia/1000
````

````python
## Lista vacia para almacenar la frecuencia acumulada
frecuencia_acum=[]

#Loop para calcular la frecuencia acumulada en cada entrada del vector
for i in range(len(frecuencia_rel)):
    frecuencia_acum.append(np.sum(frecuencia_rel[0:i]))
````

````python
# Creamos figura y establecemos tamaño
plt.figure(figsize=(14, 8), dpi=300)

# Codigo del grafico de linea y color
plt.plot(valores, frecuencia_acum)


#Titulo del grafico
plt.title('Distribución acumulada')

#Titulo del eje x
plt.xlabel('Numero de sobres', size=16)

#Titulo del eje y
plt.ylabel('Probabilidad', size=16)

##valores ejes
plt.xticks(size=12)
plt.yticks(size=12)

#Imprime el grafico
plt.show()
````

### Interpretación de la curva: 
La curva anterior muestra la distribución de probabilidad acumulada del número total de sobres necesarios para completar el álbum Panini. El eje X representa el número de sobres que se tuvieron que abrir para llenar el album en cada simulación. El eje Y muestra la probabilidad acumulada de que el álbum esté lleno después de abrir cierto número de sobres. 

En particular, la curva es creciente en todo su dominio y presenta una pendiente variable. Esto indica que, a medida que el número de sobres aumenta, la probabilidad de completar el álbum con X o menos sobres también incrementa, convergiendo al 100% cuando se destapan más de 1600 sobres. 

Se puede ver que el la probabilidad de llenar el album con menos de 600 sobres es cercana a 0. Entre 700 y 1200 sobres la probabilidad acumulada de llenar el album aumenta rápidamente. Entre los 1200 y 1600 sobres, la probabilidad de llenar el album sigue aumentando pero la pendiente de la curva es menor. Después de 1600 sobres la probabilidad de haber llenado el album es del 100%.
